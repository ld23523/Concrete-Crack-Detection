```python
import kagglehub
import shutil
import os
```

```python
# Download Crack Concrete Images for Classification dataset
def download_kaggle_dataset(path):
    print("Downloading dataset using KaggleHub...")

    print(f"Downloaded to: {path}")

    # Move into your project structure
    target_dir = "../data/raw/concrete_crack"

    if not os.path.exists(target_dir):
        shutil.copytree(path, target_dir)

    print(f"Dataset copied to: {target_dir}")
```

```python
# Download Crack Concrete Images for Classification dataset
download_kaggle_dataset(path = kagglehub.dataset_download(
        "arnavr10880/concrete-crack-images-for-classification"
))
```

```python
# Download Crackforest dataset
download_kaggle_dataset(path = kagglehub.dataset_download(
        "mahendrachouhanml/crackforest"
))
```
